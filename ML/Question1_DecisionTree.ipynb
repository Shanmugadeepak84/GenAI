```python
#Step1 -import required libraries
from sklearn.datasets import load_diabetes
import pandas as pd

# Load the diabetes dataset
diabetes = load_diabetes()

# Convert to pandas DataFrame
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)

# Add target column
df['target'] = diabetes.target

# Display the first few rows
#print(df.head())

df

```

```python
target=df['target']
target
```

```python
features=df.drop('target',axis=1)
features
```

```python
#Step3: Split tarining data and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.75,random_state=100)
print(x_train)
```

```python
#Step4: Fit the model
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier(criterion="entropy",splitter='best',max_depth=5) #instantiating the class DecisionTreeClassifier
dec.fit(x_train,y_train)
```

```python
# PLOT OF TREE
from sklearn import tree
import matplotlib.pyplot as plt
tree.plot_tree(dec)
plt.show()
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
y_pred = dec.predict(x_test)  # Only predict on test set
y_pred.shape
```

```python
from sklearn.metrics import accuracy_score
```

```python
#Step5: Calculate accuracy score
accuracy_score(y_test, y_pred)
```
