```python
# zephyr_gguf_cpu.py
from huggingface_hub import hf_hub_download
from llama_cpp import Llama
import os, multiprocessing

# 1) Download a Zephyr GGUF (4-bit is a good start)
REPO_ID = "TheBloke/zephyr-7B-beta-GGUF"
FILENAME = "zephyr-7b-beta.Q4_K_M.gguf"   # you can try Q5_K_M / Q6_K if you have more RAM

gguf_path = hf_hub_download(repo_id=REPO_ID, filename=FILENAME)
print("Model:", gguf_path)

# 2) Load the model
llm = Llama(
    model_path=gguf_path,
    n_ctx=4096,                           # context size
    n_threads=max(1, multiprocessing.cpu_count() - 1),  # use most CPU cores
    # n_gpu_layers=0  # (CPU only)
)

# 3) Zephyr chat-style prompt (system/user/assistant roles)
prompt = (
    "<|system|>\nYou are a concise, helpful assistant.\n</s>\n"
    "<|user|>\nWrite a short friendly greeting and share one cool fact about octopuses.\n</s>\n"
    "<|assistant|>\n"
)

# 4) Generate
out = llm(
    prompt=prompt,
    max_tokens=200,
    temperature=0.8,
    top_p=0.95,
    repeat_penalty=1.1,
    stop=["</s>", "<|user|>"],            # stop at end of answer / next turn
)

print(out["choices"][0]["text"].strip())
```
