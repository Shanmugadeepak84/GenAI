```python
# =======================
# 1. Imports & Config
# =======================
import os
import glob
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from PIL import Image

IMG_SIZE = 256
BATCH_SIZE = 16
EPOCHS = 25
NUM_CLASSES = 2   # human vs background


# =======================
# 2. Locate Data
# =======================
# Adjust if Kaggle unzipped to another folder
import glob
import os

DATASET_DIR = "dataset"

# Find candidate images (jpg, jpeg, png) but exclude "trimap"/mask files
images = sorted(glob.glob(os.path.join(DATASET_DIR, "*.jpg"))) \
       + sorted(glob.glob(os.path.join(DATASET_DIR, "*.jpeg"))) \
       + sorted(glob.glob(os.path.join(DATASET_DIR, "*.png")))

# Remove anything that looks like a mask (contains "mask" or "trimap")
images = [f for f in images if "mask" not in f.lower() and "trimap" not in f.lower()]

# Collect masks (assuming *_mask.png naming or folder-based)
masks = [f for f in images if f.endswith(".png") or "mask" in f.lower()]

print("Images found:", len(images))
print("Masks found :", len(masks))
print("Sample image:", images[:3])
print("Sample mask :", masks[:3])




# =======================
# 3. Preprocessing
# =======================
def load_image_mask(img_path, mask_path):
    # Load image
    img = Image.open(img_path).convert("RGB").resize((IMG_SIZE, IMG_SIZE))
    mask = Image.open(mask_path).convert("L").resize((IMG_SIZE, IMG_SIZE))

    img = np.array(img) / 255.0
    mask = np.array(mask)

    # Ensure mask is binary {0,1}
    mask = np.where(mask > 127, 1, 0).astype(np.uint8)
    mask = np.expand_dims(mask, axis=-1)  # (H,W,1)

    return img, mask

# Quick test
sample_img, sample_mask = load_image_mask(images[0], masks[0])
plt.subplot(1,2,1); plt.imshow(sample_img); plt.title("Image")
plt.subplot(1,2,2); plt.imshow(sample_mask[:,:,0]); plt.title("Mask")
plt.show()


# =======================
# 4. Dataset pipeline
# =======================
X = []
Y = []
for i, (img_path, mask_path) in enumerate(zip(images, masks)):
    img, mask = load_image_mask(img_path, mask_path)
    X.append(img)
    Y.append(mask)

X = np.array(X, dtype=np.float32)
Y = np.array(Y, dtype=np.uint8)

print("X:", X.shape, "Y:", Y.shape)

X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)

train_ds = tf.data.Dataset.from_tensor_slices((X_train, Y_train)).shuffle(500).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_ds   = tf.data.Dataset.from_tensor_slices((X_val, Y_val)).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)


# =======================
# 5. U-Net Model
# =======================
def conv_block(x, filters):
    x = keras.layers.Conv2D(filters, 3, padding="same", activation="relu")(x)
    x = keras.layers.Conv2D(filters, 3, padding="same", activation="relu")(x)
    return x

def encoder_block(x, filters):
    c = conv_block(x, filters)
    p = keras.layers.MaxPooling2D((2,2))(c)
    return c, p

def decoder_block(x, skip, filters):
    us = keras.layers.UpSampling2D((2,2))(x)
    concat = keras.layers.Concatenate()([us, skip])
    c = conv_block(concat, filters)
    return c

def build_unet(input_shape=(IMG_SIZE, IMG_SIZE, 3)):
    inputs = keras.layers.Input(input_shape)

    # Encoder
    c1, p1 = encoder_block(inputs, 64)
    c2, p2 = encoder_block(p1, 128)
    c3, p3 = encoder_block(p2, 256)
    c4, p4 = encoder_block(p3, 512)

    # Bridge
    bn = conv_block(p4, 1024)

    # Decoder
    d1 = decoder_block(bn, c4, 512)
    d2 = decoder_block(d1, c3, 256)
    d3 = decoder_block(d2, c2, 128)
    d4 = decoder_block(d3, c1, 64)

    outputs = keras.layers.Conv2D(NUM_CLASSES, 1, activation="softmax")(d4)

    return keras.Model(inputs, outputs)

model = build_unet()
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
model.summary()


# =======================
# 6. Train
# =======================
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS
)


# =======================
# 7. Visualize Predictions
# =======================
def display_sample(X, Y, preds=None):
    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1); plt.imshow(X); plt.title("Image")
    plt.subplot(1,3,2); plt.imshow(Y[:,:,0]); plt.title("Mask")
    if preds is not None:
        plt.subplot(1,3,3); plt.imshow(np.argmax(preds, axis=-1)); plt.title("Predicted")
    plt.show()

# Test on validation sample
for imgs, masks in val_ds.take(1):
    preds = model.predict(imgs)
    display_sample(imgs[0].numpy(), masks[0].numpy(), preds[0])
```
